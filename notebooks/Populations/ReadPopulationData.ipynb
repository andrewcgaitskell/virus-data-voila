```python
##def poptable(mf_in):
path2file =  '/home/jupyterlab/data/data.csv'
if os.path.exists(path2file):
  os.remove(path2file)
else:
  print("The file does not exist")

mf_in = 'male'
malepersons_melted.to_csv(path2file)
table = 'raw_pop_malepersons'
sqlcmnd = 'COPY "' + table + '" FROM \''+ path2file + '\' DELIMITER \',\' CSV;'
with engine.connect().execution_options(autocommit=True) as con:
    con.execute(sqlcmnd)

if os.path.exists(path2file):
    os.remove(path2file)
else:
    print("The file does not exist")
    
mf_in = 'female'
femalepersons_melted.to_csv(path2file)
table = 'raw_pop_femalepersons'
sqlcmnd = 'COPY "' + table + '" FROM \''+ path2file + '\' DELIMITER \',\' CSV;'
with engine.connect().execution_options(autocommit=True) as con:
    con.execute(sqlcmnd)

if os.path.exists(path2file):
    os.remove(path2file)
else:
    print("The file does not exist")

allpersons_melted.to_csv(path2file)
table = 'raw_pop_allpersons'
sqlcmnd = 'COPY "' + table + '" FROM \''+ path2file + '\' DELIMITER \',\' CSV;'
with engine.connect().execution_options(autocommit=True) as con:
    con.execute(sqlcmnd)


if os.path.exists(path2file):
    os.remove(path2file)
else:
    print("The file does not exist")
```
